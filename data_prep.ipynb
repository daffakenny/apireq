```python
# Imports
import numpy as np
import pandas as pd

# Dataviz
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# import warnings
# warnings.filterwarnings("ignore")
```

```python
# Season 2021 = 2021/2022

# Standings of the whole league, with each team's id and name
nbastandings = pd.read_csv("nbastandings_normalized.csv")

with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(nbastandings)
```

```python
# Dropping unused columns on standings
nbastandings.drop(nbastandings.columns[[0, 1, 2, 4, 5, 6, 9, 11, 16, 17, 18, 19, 20, 25, 30]], axis=1, inplace = True)
nbastandings
```

```python
# Sorting by Conference Rank since there are 2 confs within the NBA
nbastandings = nbastandings.sort_values(by=['conference.rank'], ascending=True)
nbastandings
```

```python
# Importing NBA Teams list
nbateams = pd.read_csv("nbateams_normalized.csv")
nbateams
```

```python
# Dropping unused columns on NBA Teams list
nbateams.drop(nbateams.columns[[0, 3, 5, 6, 7, 8, 11, 12, 13, 14, 15, 16]], axis=1, inplace = True)
nbateams
```

```python
# Importing each team's stats within the regular season
nbateamstats = pd.read_csv("nbateamstats_normalized.csv")
nbateamstats
```

```python
# Create dictionary to map teamID/id to teamName/name

teamID2name = dict(zip(nbateams['id'], nbateams['name']))
print(teamID2name)
```

```python
# Adding team name to their respective id using dict
nbateamstats['name'] = nbateamstats['teamID'].map(teamID2name)
nbateamstats
```

```python
# Move team name to [1]

nbateamstats.insert(1, 'name', nbateamstats.pop('name'))
nbateamstats
```

```python
 # Column renaming for consistensy
nbastandings.rename(columns={'team.id': 'teamID'
                            , 'team.name': 'teamName'}, inplace=True)
nbastandings
```

```python
nbastandings['teamName'] = nbastandings['teamID'].map(teamID2name)

col_interest = ['conference.name'
                , 'conference.rank'
                , 'conference.win'
                , 'conference.loss'
                , 'win.home'
                , 'win.away'
                , 'loss.home'
                , 'loss.away']
nbateamstats[col_interest] = nbastandings[col_interest]
```

```python
nbateamstats
```

```python
nbateamstats.to_csv('nbateamstats_final.csv', index=False)
```

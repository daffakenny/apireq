```python
# Imports
import numpy as np
import pandas as pd

# Dataviz
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# import warnings
# warnings.filterwarnings("ignore")
```

```python
# Season 2021 = 2021/2022

# Standings of the whole league, with each team's id and name
nbastandings = pd.read_csv("nbastandings_normalized.csv")

with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    display(nbastandings)
```

```python
# Dropping unused columns on standings
nbastandings.drop(nbastandings.columns[[0, 1, 2, 4, 5, 6, 9, 11, 16, 17, 18, 19, 20, 25, 30]], axis=1, inplace = True)
nbastandings
```

```python
# Column renaming for consistensy
nbastandings.rename(columns={'team.id': 'teamID'
                             , 'team.name': 'teamName'
                             , 'team.code': 'teamCode'
                             , 'conference.name': 'conferenceName'
                             , 'conference.rank': 'conferenceRank'
                             , 'conference.win': 'conferenceWin'
                             , 'conference.loss': 'conferenceLoss'
                             , 'win.home': 'winHome'
                             , 'win.away': 'winAway'
                             , 'loss.home': 'lossHome'
                             , 'loss.away': 'lossAway'
                             , 'team.code': 'teamCode'}, inplace=True)
nbastandings
```

```python
# Importing each team's stats within the regular season
nbateamstats = pd.read_csv("nbateamstats_normalized.csv")
nbateamstats
```

```python
# Create dictionary to map ID, Name, and Code

team_info_dict = {id_: {'name': name, 'code': code} for 
                  id_, name, code in zip(nbastandings['teamID'], nbastandings['teamName'], nbastandings['teamCode'])}
print(team_info_dict)
```

```python
nbateamstats = nbateamstats.merge(nbastandings[['teamID'
                                                , 'teamName'
                                                , 'teamCode'
                                                , 'conferenceName'
                                                , 'conferenceRank'
                                                , 'conferenceWin'
                                                , 'conferenceLoss'
                                                , 'winHome'
                                                , 'winAway'
                                                , 'lossHome'
                                                , 'lossAway']], on='teamID', how='right')

with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    display(nbateamstats)
```

```python
# Move columns

nbateamstats.insert(1, 'teamName', nbateamstats.pop('teamName'))
nbateamstats.insert(2, 'teamCode', nbateamstats.pop('teamCode'))

with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    display(nbateamstats)
```

```python
nbateamstats.to_csv('nbateamstats_final.csv', index=False)
```
